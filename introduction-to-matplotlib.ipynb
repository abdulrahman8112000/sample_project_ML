# matplotlib example workflow

```python
#0. import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

#1. Prepeare data
x=[1,2,3,4]
y=[11,22,33,44]

#2. setup plot
fig,ax=plt.subplots(figsize=(10,10))

#3. plot data
ax.plot(x,y)

#4. customize plot
ax.set(title="Simple plot",
      xlabel="x axis",
      ylabel="y axis")

#5. save and show fig
fig.savefig("images/sample-plot.png")
```

## making figures with numpy arrays

```python
x=np.linspace(0,10,100)
```

```python
fig,ax=plt.subplots()
ax.scatter(x,np.sin(x))
```

```python
nut_butter_sales={
    "ALmon butter":10,
    "peanut butter":8,
    "cashew butter":12
}

fig,ax=plt.subplots()
ax.bar(nut_butter_sales.keys(),nut_butter_sales.values());
```
