# matplotlib example workflow

```python
#0. import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

#1. Prepeare data
x=[1,2,3,4]
y=[11,22,33,44]

#2. setup plot
fig,ax=plt.subplots(figsize=(10,10))

#3. plot data
ax.plot(x,y)

#4. customize plot
ax.set(title="Simple plot",
      xlabel="x axis",
      ylabel="y axis")

#5. save and show fig
fig.savefig("images/sample-plot.png")
```

```python
hd=pd.read_csv("heart-disease.csv")
hd.plot(kind="bar");
```

```python
hd.shape
```

```python
hd.ndim
```

```python
over_50=hd[hd["age"]>50]
over_50
```

```python
#pyplot method

hd.plot(x="chol",y="age",kind="scatter")
```

```python
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=hd["age"],y=hd["chol"],c=hd["target"]);
ax.set(title="heart disease and cholestrol levels",xlabel="Age",ylabel="cholestrol");
ax.legend(*scatter.legend_elements(),title="Target")
ax.axhline(over_50["chol"].mean(),linestyle="--");
```

```python
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10))
scatter=ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])
ax0.set(title="Heart disease and cholestrol levels",xlabel="Age",ylabel="Cholestrol");
ax0.legend(*scatter.legend_elements(),title="target")
ax0.axhline(over_50["chol"].mean(),linestyle="--");

scatter2=ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"])
ax1.set(title="heart disease and thalach",xlabel="AGE",ylabel="thalach")
```

```python
plt.style.use('seaborn-whitegrid')
over_50["age"].plot()
```

```python
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10))
scatter=ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"],cmap="winter")
ax0.set(title="Heart disease and cholestrol levels",xlabel="Age",ylabel="Cholestrol");
ax0.set_xlim([50,80])
ax0.legend(*scatter.legend_elements(),title="target")
ax0.axhline(over_50["chol"].mean(),linestyle="--");

scatter2=ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"],cmap="winter")
ax1.set(title="heart disease and thalach",xlabel="AGE",ylabel="thalach");
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])
```
